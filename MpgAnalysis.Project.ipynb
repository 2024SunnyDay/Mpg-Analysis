```python
import seaborn as sns
```

```python
mpg=sns.load_dataset("mpg")
mpg
```

```python
Create a visual showing the relationship between horsepower and fuel economy.
Create a table summarizing fuel economy by country of origin
Create a visual showing how fuel economy has changed over time
Create two other visuals or tables related to fuel economy.
```

```python
#A visual showing horsepower and fuel economy relationship
#the more horsepower used = more gas per gallon used
sns.relplot(
            data=mpg,
            x="mpg", y="horsepower")
```

```python
#A table summarizing fuel economy by country of origin
#USA is using less fuel per gallon, than Japan and Europe. Japan burning more fuel than both Europe and the USA. 
sns.relplot(data=mpg,
           x="origin",y="mpg")
```

```python
sns.displot(data=mpg, x="mpg", hue="model_year")
#Showing how the economy of fuel has changed over time.
```

```python
mpg.describe()
#here you can see at different levels the effects of horsepower and the miles per gallon 
```

```python
sns.catplot( data=mpg, x="model_year",y="horsepower",kind="bar")
#With this chart you can see which model_year used the most horsepower.
```
